# 🧹 第1章：ZOZO公式IRデータのクレンジングとCSVエクスポート
💡 **分析意図:** ZOZO公式の決算資料から抽出した信頼性の高いリアルデータを定義し、欠損値のない完璧なデータパイプラインを構築してローカル環境にCSVとして書き出すための準備セクションです。

```python
# ─────────────────────────────────────────────────────────────────────
# 📦 Cell 1: 必要な外部ライブラリのインポート
# ─────────────────────────────────────────────────────────────────────
import pandas as pd
import numpy as np

print("✅ [INFO] データ前処理用ライブラリのインポートが正常に完了しました。")
```

```python
# ─────────────────────────────────────────────────────────────────────
# 📦 Cell 2: ZOZO公式リアルデータの定義 ＆ 単価自動計算
# ─────────────────────────────────────────────────────────────────────
data = {
    'year': ['2025', '2025', '2025', '2025', '2026', '2026', '2026', '2026'],
    'quarter': ['25/03 1Q', '25/03 2Q', '25/03 3Q', '25/03 4Q',
                '26/03 1Q', '26/03 2Q', '26/03 3Q', '26/03 4Q'],
    'active_buyers': [11790269, 11870844, 12057726, 12217038,
                      12365080, 12529665, 12809389, 13173445],
    'annual_spend_per_user': [42947, 43171, 43307, 42953,
                              42861, 42404, 41771, 41323],
    'items_purchased_per_user': [10.9, 11.0, 11.0, 10.9,
                                 10.8, 10.7, 10.6, 10.6]
}

df = pd.DataFrame(data)
df['purchase_price'] = (df['annual_spend_per_user'] / df['items_purchased_per_user']).round(0).astype(int)

# 🎯 【自動検証】データが壊れていないか厳密にチェック（改善提案4）
assert len(df) == 8, "[ERROR] 元データの行数が不正です。"
assert df['purchase_price'].iloc[0] == 3940, "[ERROR] 単価の計算ロジックに不整合があります。"

print("✅ [SUCCESS] ZOZO公式データのクレンジングと自動検証が100%成功しました。")
```

```python
# ─────────────────────────────────────────────────────────────────────
# 📦 Cell 3: dataset/zozo_ir_data.csv への書き出し
# ─────────────────────────────────────────────────────────────────────
import os

os.makedirs('dataset', exist_ok=True)
df.to_csv('dataset/zozo_ir_data.csv', index=False, encoding='utf-8')

print("✅ [SUCCESS] dataset/zozo_ir_data.csv に実データを正常にエクスポートしました！")
```
